Script: Perform interpolation on for missing data.
We will use the pandas interpolate() function to perform the interpolation.

```python
import os
import pandas as pd
import numpy as np
from math import floor, ceil

from sensorfabric.athena import athena
import time
from datetime import datetime, timedelta

%set_env AWS_PROFILE=uaprofile

db = athena(database='elise', offlineCache=True)
```

```python
# Total PID 56 - PID 123 excluded from the analysis due to a lot of missing data at regular times 
# of the day.
pids = [1,2,7,10,11,15,29,35,36,48,55,61,67,69,72,73,75,76,77,80,88,90,91,93,97,
        99,102,104,106,107,108,110,111,113,120,124,126,129,136,138,141,144,152,
        156,161,169,170,173,179,183,189,191,193,195,197]

print(len(pids))
```

```python
exp_dir = '/home/chinmai/src/Oura/Data/No_ini_nonwear_data_y/'

for pid in pids:
    print('Processing PID: ',pid)
    # Read the temperature data from the CSV file in the exp_dir
    fname = os.path.join(exp_dir,str(pid)+'_5temp_y.csv')
    print (fname)
    # Read temperature data from the file.
    data = np.loadtxt(fname, delimiter=',')

    # Exclude the y values - The last column
    # Include only 21 days of data
    y_data = data[:,-1:]
    x_data = data[:,:-1]
   
    # Mask all values that are missing (-1) and non-wear(-2)
    x_data = np.ma.masked_equal(x_data, -2)
    x_data = np.ma.masked_equal(x_data, -1)
    r,c = x_data.shape
    #print (x_data.shape)
    # Reshape the temperature data to a 1D array
    x_temp = np.reshape(x_data,(r*c,))
    #print(x_temp.shape)

    # We will be using pandas library, whihc in turn calls scipy library for interpolation calls
    df = pd.Series(x_temp)
    df_linear = df.interpolate(method='linear')
    x_lin = df_linear.to_numpy(dtype='float32')
    x_lin = np.reshape(x_lin,(r,c))
    print (x_lin.shape)
    d_lin = np.concatenate((x_lin,y_data),axis=1)
    v = np.savetxt(str(pid)+'_5temp_linIP_y.csv',d_lin,delimiter=',',fmt='%.4f')
    
```
